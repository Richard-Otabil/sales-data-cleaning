```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\OTABIL\Downloads\python\netflix_titles.csv", encoding="latin1", usecols=range(12))
print(df.head())
print(df.shape)
print(df.isnull().sum())
```

```python
df["director"] = df["director"].fillna("No Director Listed")
df["cast"] = df["cast"].fillna("No Cast Listed")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Not Rated")
df["duration"] = df["duration"].fillna("Unknown")
df["date_added"] = df["date_added"].fillna("Unknown")

print(df.isnull().sum())
```

```python
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
# Standardize to lowercase, strip extra spaces
df["type"] = df["type"].str.strip().str.lower()
df["rating"] = df["rating"].str.strip().str.lower()
df["country"] = df["country"].str.strip().str.title()

# Confirm
print(df["type"].unique())
print(df["rating"].unique())
```

```python
# Fix misplaced duration values in rating column
wrong_ratings = ['74 min', '84 min', '66 min']
df["rating"] = df["rating"].replace(wrong_ratings, "not rated")

# Confirm
print(df["rating"].unique())
```

```python
# Unify similar rating values
df["rating"] = df["rating"].replace({'nr': 'not rated', 'ur': 'not rated'})

# Confirm
print(df["rating"].unique())
```

```python
# Check current columns
print(df.columns.tolist())

# Rename to clean, lowercase, no spaces
df.columns = ["show_id", "type", "title", "director", "cast", "country", 
              "date_added", "release_year", "rating", "duration", 
              "listed_in", "description"]

# Confirm
print(df.columns.tolist())
```

```python
print(df.dtypes)
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
print(df.dtypes)
print()
print(df["date_added"].head())
```

```python
df.to_csv(r"C:\Users\OTABIL\Downloads\python\netflix_titles_cleaned.csv", index=False)
print("Cleaned file saved!")
```
